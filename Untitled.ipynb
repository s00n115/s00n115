```python
menu = {}
print("Welcome to the Menu-Based Program!\n Please select an option:")
menu['1']="print pattern" 
menu['2']="Rotate Array"
menu['3']="Help"
menu['4']="Exit"

while True: 
    options=sorted(menu.keys())
    
    for entry in options: 
        print (entry, menu[entry])

    selection=input("Please Select:") 
    if selection =='1': 
        n=int(input("Enter the number of rows for the pattern : "))
        for i in range (0,n):
            for j in range (n,i,-1):
                print("* ", end="")
            print()
        print("\n ")
    elif selection == '2': 
        def rotate_array(n, k, arr):
            if n < 0 or k < 0 or k >= n:
                raise ValueError("Invalid input.")
            for i in range(k):
                temp = arr[n - 1]
                for j in range(n - 1, 0, -1):
                    arr[j] = arr[j - 1]
                    arr[0] = temp
            return arr
        n = int(input("Enter the number of elements (n) : "))
        k = int(input("Enter the number of steps (k): "))
        arr = list(map(int, input("Enter the array elements separated by space : ").split()))
        rotated_array = rotate_array(n, k, arr)
        print("The rotated array is:", rotated_array)
        print("\n ")
    elif selection == '3':
        def help_message():
            print("--- Help ---")
            print("1. print a pattern with 'n' row of decreasing asterists.")
            print("2. Rotate an array of 'n' element to the right by 'k' steps.")
            print("3. Display this help message")
            print("4. Exit the program")
        help_message()
        print("\n ")
    elif selection == '4': 
        print("Exiting the program.GOODBYE!")
        break
    else: 
        print ("Unknown Option Selected!") 
```
